```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df_survey = pd.read_csv('https://cocl.us/datascience_survey_data', index_col=0)
```

```python
df_survey1 = df_survey.astype(float)
print(df_survey1.dtypes)
```

```python
df_survey1.sort_values(['Very interested'], ascending=False,inplace=True)
df_surveypercent = (df_survey1/2233)
df_surveypercent2 = df_surveypercent.round(decimals=2)
df_surveypercent2
```

```python
ax1 = df_surveypercent2.plot(kind = 'bar',
                figsize = (20,8),
                width = 0.8,
                color = ('#5cb85c', '#5bc0de', '#d9534f'),
                fontsize = 14)
ax1.set_title("Percentage of Respondents' Interest in Data Science", 
             fontsize = 16)
ax1.legend(fontsize = 14)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)

# Adding the percentage values    
for p in ax1.patches:
    ax1.annotate("{:.2%}".format(p.get_height()),
                xy=(p.get_x()+0.02, p.get_height()+0.01))
```
